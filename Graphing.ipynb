```python
import pandas as pd
```

```python
data = pd.read_csv("Data_2.csv",error_bad_lines=False)
```

```python
data_title = data["Document Title"]
```

```python
data_keywords = data["Author Keywords"]
```

```python
data_keywords = data_keywords.dropna()
```

```python
len(data_keywords)
```

```python
data_title = data_title.dropna()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,1), stop_words = 'english')
X = vectorizer.fit_transform(data_keywords)

names = vectorizer.get_feature_names() # This are the entity names (i.e. keywords)
new_df = pd.DataFrame(data = X.toarray(), columns = names)
```

```python
from sklearn.metrics.pairwise import pairwise_distances
jac_sim = 1 - pairwise_distances(new_df.T, metric = "jaccard")
# optionally convert it to a DataFrame
jac_sim = pd.DataFrame(jac_sim, index=new_df.columns, columns=new_df.columns)
```
